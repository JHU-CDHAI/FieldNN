# Change PID and ECID

```python
import os

folder = 'data/Simulations/'
file_list = [os.path.join(folder, i) for i in os.listdir(folder) if '.csv' in i and 'Encounter' not in i]
file_list
```

```python
col = 'PID'
df_list = [pd.read_csv(file) for file in file_list]
ID_series = pd.concat([df[col] for df in df_list if col in df.columns])
# {row[col]: idx for idx, row in ID_series.value_counts().reset_index().iterrows()}
# ID_series
PID_2_new = {row['index']: idx for idx, row in ID_series.value_counts().reset_index().iterrows()}
# ID_series
len(PID_2_new)
```

```python
PID_2_new
```

```python
col = 'ECID'
df_list = [pd.read_csv(file) for file in file_list]
ID_series = pd.concat([df[col] for df in df_list if col in df.columns])
# {row[col]: idx for idx, row in ID_series.value_counts().reset_index().iterrows()}
# ID_series
ECID_2_new = {row['index']: idx for idx, row in ID_series.value_counts().reset_index().iterrows()}
# ID_series
len(ECID_2_new)
```

```python
# ECID_2_new
```

```python
# ID_series.value_counts()
```

```python
for filepath in file_list:
    print(filepath)
    df = pd.read_csv(filepath)
    df['PID'] = df['PID'].apply(lambda x: 'P' + str(PID_2_new[x]))
    if 'ECID' in df.columns:
        df['ECID'] = df['ECID'].apply(lambda x: 'EC' + str(ECID_2_new[x]))
        
    if 'DT' in df.columns: 
        df = df.sort_values(['PID', 'DT']).reset_index(drop = True)
    else:
        df = df.sort_values('PID').reset_index(drop = True)
    df.to_csv(filepath, index = False)
```

# Encounter Information

```python
import os

folder = 'data/Simulations/'
file_list = [os.path.join(folder, i) for i in os.listdir(folder) if '.csv' in i]
file_list
```

```python
df_list = [pd.read_csv(file) for file in file_list]


a = pd.concat([df[['PID', 'ECID', 'DT']]  for df in df_list if 'ECID' in df.columns and 'DT' in df.columns])
a['DT'] = pd.to_datetime(a['DT'])
a = a.groupby('ECID').apply(lambda x:{'PID': x['PID'].max(), 
                                      'ECID': x['ECID'].max(), 
                                      'DT_min': x['DT'].min(), 
                                      'DT_max': x['DT'].max(), 
                                      'BasicInfo': '&'.join([random.choice([i for i in 'ABCD']), 
                                                          random.choice([i for i in 'WXYZ'])]),
                                     }).to_list()

df_ECID = pd.DataFrame(a).sort_values(['PID', 'DT_min']).reset_index(drop = True)
df_ECID
```

```python
RecName = 'EC'
path = f'data/Simulations/{RecName}.csv'
print(path)

df_ECID.to_csv(path, index = False)
```
